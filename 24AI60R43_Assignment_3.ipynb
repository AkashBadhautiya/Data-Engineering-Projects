```python
# Importing  the  module required  for handling the ZIP files
from zipfile import ZipFile

#using with,  Opening the ZIP file with name 'Document.zip' as read mode
with ZipFile('Document.zip', 'r') as document_zip:
    # Extracting the  all the files of the ZIP file to the current directory
    document_zip.extractall()


```

```python
# Importing  the glob module to work with file patterns
import glob

# Defining the pattern to match all files and storing all the files  in the 'doc' directory
document_files = glob.glob("doc/*")

# Printing the total number of files exits  in the 'doc' directory
print(len(document_files))
```

```python
def calculate_word_probabilities(word_frequencies):
    # Calculating the total number of words by adding  all the  frequencies of all word
    total_word_count = sum(word_frequencies.values())

    # Initializing an empty dictionary to store the probabilities  of each word
    word_probabilities = {}

    # Calculating  the probability of each word in file
    for word, frequency in word_frequencies.items():
        # Applying Laplace smoothing as given in the assignment and storing the result in the dictionary
        word_probabilities[word] = (frequency + 1) / (total_word_count + 2)

    # Returning the dictionary containing the  probabilities of word
    return word_probabilities
```

```python
import re #importing the regular expression(re)

# Initializing the  dictionary to store word probability for each file
file_word_probabilities = {}

# Initializing a set data structure for storing the unique vocabulary across all files
unique_vocabulary = set()

# Iterating through each file in the document_files
for file_path in document_files:

    # using 'with'  ,Opening  and reading  the content of the file
    with open(file_path, 'r') as file_content:
        text = file_content.read().lower()  # Converting the  text of a file to lowercase

        # Removing the  punctuation and numbers from the text
        text = re.sub(r'[^\w\s]', '', text)  # Removing the punctuation
        text = re.sub(r'[0-9]', '', text)    # Removing the  numbers

        # Spliting the text into individual words in file
        word_list = text.split()

        # Updating  the vocabulary set with the words from the current file
        unique_vocabulary.update(word_list)

        # Extracting  the file number from the file name
        file_number = file_path.split('_')[-1].split('.')[0]

        # Calculating the  word frequencies in the current file
        word_frequencies = {} # initializing the dictionary for storing the probability of each word
        for word in word_list:
            if word in word_frequencies:
                word_frequencies[word] += 1 # incrementing the word frequency is alread indexed
            else:
                word_frequencies[word] = 1 # assigning the frequency to new word

        # Storing the word probabilities for the current file number as a dictionary
        file_word_probabilities[file_number] = calculate_word_probabilities(word_frequencies)

# Printing  the dictionary containing word probabilities for each file
print(file_word_probabilities)
```

```python

# Initializing a dictionary which  stores the word probabilities for the query document
query_word_probabilities = {}

# Initializing  a set data structure for storing the unique vocabulary from the query document
query_vocabulary = set()

#using 'with' , Opening and reading the content of the query document
with open('Query.txt', 'r') as query_file:
    query_text = query_file.read().lower()  # Convertin the  text of query document to lowercase

    # Removing the  punctuation and numbers from the query document
    query_text = re.sub(r'[^\w\s]', '', query_text)  # Removing the  punctuations
    query_text = re.sub(r'[0-9]', '', query_text)    # Removing the numbers

    # Spliting the cleaned document(text) into individual words
    query_words = query_text.split()

    # Updating the vocabulary set with the words in the query document
    query_vocabulary.update(query_words)

    # Calculating the word frequency in the query document
    query_word_frequencies = {}
    for word in query_words:
        if word in query_word_frequencies:
            query_word_frequencies[word] += 1 # increamenting the frequency of word if indexed
        else:
            query_word_frequencies[word] = 1 #assigning the frequency of new word

    # Storing  the calculated probability of word for the query document
    query_word_probabilities = calculate_word_probabilities(query_word_frequencies)

# Printing the dictionary which containing word probabilities for  query document
print(query_word_probabilities)
```

```python
import math

def calculate_similarity_score(doc_word_probabilities, query_word_probabilities):
    """
    Calculating the similarity score between a document file in files  and a query document using the Kullback-Leibler divergence as mentioned in  Assignment.
    """
    similarity_score = 0.0  # Initializing the similarity_score variable of type float  to zero

    # Iterating   all words in the query_word-probabilities dictionary
    for word, query_prob in query_word_probabilities.items():
        if word in doc_word_probabilities:
            # Retrieving the probability of the word in the doc_word_probabilities
            doc_prob = doc_word_probabilities[word]

            # Ensuring the  probabilities  of word from both document are positive, for  avoid log(0) which is not defined
            if doc_prob > 0 and query_prob > 0:
                # Updating  the similarity score using formula given the assignment document
                similarity_score += doc_prob * math.log(query_prob / doc_prob)

    return similarity_score
```

```python
# Initializing the  variables for getting the most similar file and its similarity score
most_similar_file = None #initialized to None
highest_similarity_score = float('-inf') # initialized to negative infinity

# Iterate  all files in the probability dictionary
for file_number, doc_probabilities in file_word_probabilities.items():

    # Calculating the similarity score between the current document file and the query file
    similarity_score = calculate_similarity_score(doc_probabilities, query_word_probabilities)

    # Updating  the most similar file and and its score, if the current similarity is higher than the highest similarity
    if similarity_score > highest_similarity_score:
        highest_similarity_score = similarity_score
        most_similar_file = file_number  # file no which is the most similar to query document

# Printin the most similar file number with its similarity score
print("The most similar document number is:", most_similar_file)
print("The similarity score of this document with query document is:", highest_similarity_score)
```
