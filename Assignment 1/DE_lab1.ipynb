## 2.7: Tuple:

A tuple is a collection which is ordered and unchangeable. In Python tuples are written with round brackets.


#### Constructing Tuples

The construction of a tuples use () with elements separated by commas. For example:

````python
tup=(1,2,3)
print(type(tup))
````

````python
# Create a tuple
t = (3,2,1)
#t[1]=10 #this is not goin to work, will give error. Tuple elements are immutable
print(type(t))
print(t)
````

````python
# Check len just like a list
len(t)
````

````python
# Use indexing just like we did in lists
#t[0]
print(t)
````

````python
# Slicing just like a list
t[1:5]
````

````python
# Can also mix object types
t = ('one',2,3,4,6,'two')

# Show
# print second to second last
t[1:4]

# [start:end:step]
# [start:end-1:step]
````

## 2.8: Sets

- Python also includes a data type for sets. A set is an unordered, unindexed , unchangeable collection with no duplicate elements.
- Basic uses include membership testing and eliminating duplicate entries.
- Set objects also support mathematical operations like union, intersection, difference, and symmetric difference.
