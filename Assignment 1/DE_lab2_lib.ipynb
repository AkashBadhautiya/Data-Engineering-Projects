# 5.2: Pandas

- Pandas is an open source Python library for data analysis.

- Pandas is built on top of the NumPy package, meaning a lot of the structure of NumPy is used or replicated in Pandas. Data in pandas is often used to feed statistical analysis in SciPy, plotting functions from Matplotlib, and machine learning algorithms in Scikit-learn.

- Pandas introduces two new data structures to Python - Series and DataFrame, both of which are built on top of NumPy (this means it's fast).

````python
# Now in order to use pandas either we should have install it if not already

# insatall pandas

# pip install pandas

# now importing pandas in order to use that

import pandas as pd
````

## Pandas Series:

A Series is a one-dimensional object similar to an array, list, or column in a table. It will assign a labeled index to each item in the Series. By default, each item will receive an index label from 0 to N, where N is the length of the Series minus one.

````python
# create a Series with an arbitrary list
s = pd.Series([7, 'Heisenberg', 3.14, -1789710578, 'Happy Eating!'])
s
````

````python
# you can explicitely define the index also

s = pd.Series([7, 'Heisenberg', 3.14, -1789710578, 'Happy Eating!'],
              index=['A', 'Z', 'C', 'Y', 'E'])
````

````python
s
````

````python
# The Series constructor can convert a dictonary as well, using the keys of the dictionary as its index.

d = {'Chicago': 1000, 'New York': 1300, 'Portland': 900, 'San Francisco': 1100,
     'Austin': 450, 'Boston': None}
cities = pd.Series(d)
cities
````

````python
# You can use the index to select specific items from the Series.

````

````python
cities[['Chicago', 'Portland', 'San Francisco']]
````

## 5.2: Pnadas Dataframe

DataFrames are an extremely powerful tool and a natural extension of the Pandas Series. By definition all a DataFrame is:

```A Pandas DataFrame consists of multiple Pandas Series that share index values.```


````python
import numpy as np
````

````python
# How to convert tabular data into dataframe

# Let's say we have

mydata = np.random.randint(0,101,(4,3))
mydata
````

````python
# index :
myindex = ['CA','NY','AZ','TX']

# columns
mycolumns = ['Jan','Feb','Mar']
````

````python
# Now convert into dataframe

df = pd.DataFrame(mydata,columns=mycolumns,index=myindex)
df
````

````python
df.sort_values('Jan')
````

````python
df1 = pd.DataFrame(mydata, columns= mycolumns, index = myindex)
df1
````

````python
df1.iloc[0] #iloc
x=df1.iloc[2,1]
print(x)
````

## 5.3: Matplotlib

Matplotlib is the library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

- Matplotlib allows you to create reproducible figures programmatically.

````python
x = np.arange(0,10)
y = 2*x
````

## Using Matplotlib with plt.plot() function calls

### Basic Matplotlib Commands

We can create a very simple line plot using the following:

````python
plt.plot(x, y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show() # Required for non-jupyter users , but also removes Out[] info
````

````python
# In order to save your image
plt.plot(x,y)
plt.savefig('example.png')
````

````python
# Coding style plots
````

````python
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7))
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend();
````

# Scatter Plots

Scatter plots can show how different features are related to one another, the main theme between all relational plot types is they display how features are interconnected to each other. There are many different types of plots that can be used to show this, so let's explore the scatterplot() as well as general seaborn parameters applicable to other plot types.

````python
# using dummy data

df = pd.read_csv("dm_office_sales.csv")
df
````

````python
df.info()
````

````python
# Scatterplot

sns.scatterplot(x='salary',y='sales',data=df)
````

````python
# Connecting with matplotlib

plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df)
````

## Seaborn Parameters

The hue and palette parameters are commonly available around many plot calls in seaborn.

### hue

Color points based off a categorical feature in the DataFrame

````python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='division')
````

````python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')
````

````python
# Exporting seaborn figure

plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)

# Call savefig in the same cell
plt.savefig('example_scatter.jpg')
````
