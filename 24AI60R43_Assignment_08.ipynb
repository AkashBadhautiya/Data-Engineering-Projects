```python
import spacy

# Load the SpaCy model
nlp = spacy.load('en_core_web_sm')

def extract_entities_and_pos(sentence):
    doc = nlp(sentence)
    # Extract named entities as a set of labels
    entities = {ent.label_ for ent in doc.ents}
    # Extract parts of speech as a list
    pos_tags = [token.pos_ for token in doc]
    return entities, pos_tags

def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union != 0 else 0

def compute_similarity_score(query, candidate):
    # Extract named entities and POS tags
    entities_q, pos_q = extract_entities_and_pos(query)
    entities_c, pos_c = extract_entities_and_pos(candidate)

    # Calculate named entity similarity score
    common_entities = len(entities_q.intersection(entities_c))
    union_entities = len(entities_q.union(entities_c))

    named_entity_score = common_entities / union_entities if union_entities != 0 else 0

    # Calculate POS score (using Jaccard similarity)
    pos_score = jaccard_similarity(set(pos_q), set(pos_c))

    # Total score
    total_score = named_entity_score + pos_score

    return {
        'common_entities': common_entities,
        'union_entities': union_entities,
        'named_entity_score': named_entity_score,
        'pos_score': pos_score,
        'total_score': total_score
    }

def find_similar_sentences(query, sentences):
    scores = {}
    for candidate in sentences:
        score_details = compute_similarity_score(query, candidate)
        scores[candidate] = score_details

    # Sort by total score in descending order
    sorted_scores = sorted(scores.items(), key=lambda x: x[1]['total_score'], reverse=True)
    return sorted_scores

#sentences
sentences = [
    "Akash is going to school today.",
    "Priya is excited to visit the art museum this weekend.",
    "Ravi is planning to go hiking in the mountains tomorrow.",
    "Sita is baking a delicious cake for her friend's birthday party.",
    "Vikram is joining a cooking class to learn new recipes.",
    "Anjali is reading a fascinating book about ancient history.",
    "Nina is practicing her dance moves for the upcoming recital.",
    "Karan is playing football with his friends at the park.",
    "Meera is studying for her final exams at the library.",
    "Arjun is attending a music concert this Friday night.",
    "Aisha is volunteering at the animal shelter this afternoon.",
    "Rahul is taking a photography class to improve his skills.",
    "Tina is visiting her grandparents over the weekend.",
    "Deepak is fixing his bicycle in the garage today.",
    "Geeta is learning to play the guitar in her free time.",
    "Neha is going for a run in the park this evening.",
    "Vishal is working on a science project for school.",
    "Lakshmi is planning a surprise party for her best friend.",
    "Amit is exploring new hiking trails this summer.",
    "Pooja is creating beautiful art for the local exhibition.",
    "Rina is organizing her closet to make room for new clothes.",
    "Rohit is studying hard to get good grades this semester.",
    "Sunita is going shopping for new shoes later today.",
    "Jatin is learning about marine biology in his class.",
    "Rita is writing a story for her creative writing assignment.",
    "Sanjay is practicing his speech for the school competition.",
    "Dev is taking his dog for a walk in the evening.",
    "Kavita is gardening and planting flowers in her backyard.",
    "Rajesh is attending a workshop on digital marketing.",
    "Nisha is training for a marathon that is happening next month.",
    "Karan is working on his art project for school.",
    "Rani is watching her favorite movie tonight.",
    "Farah is participating in a community clean-up event this weekend.",
    "Sunil is going to the gym to work out after work.",
    "Tarun is writing an essay about climate change for his class."
]


#query sentence
query_sentence = "Neha is planning to visit the park this afternoon to enjoy the beautiful weather."


# Find and print all similarity scores
all_scores = find_similar_sentences(query_sentence, sentences)

print("All Similarity Scores:")
for candidate, score_details in all_scores:
    print(f"Sentence: '{candidate}'")
    print(f"  Common Named Entities: {score_details['common_entities']}")
    print(f"  Union of Named Entities: {score_details['union_entities']}")
    print(f"  Named Entity Score: {score_details['named_entity_score']:.4f}")
    print(f"  POS Score: {score_details['pos_score']:.4f}")
    print(f"  Total Score: {score_details['total_score']:.4f}\n")

# Print top 5 sentences
top_5 = all_scores[:5]
print("Top 5 Similar Sentences:")
for candidate, score_details in top_5:
    print(f"Sentence: '{candidate}', Total Score: {score_details['total_score']:.4f}")
```
