```python
#Task 1: Lists and Tuples
#1.	Create a list of integers from 1 to 100. Slice the list to get every 5th element and print it.

list1=[i for i in range(1,100)]
print(list1[::5])

#2.	Convert the list to a tuple and print the first and last elements.
t=tuple(list1)
print(t[0],t[len(t)-1])

#3.	Create a nested list comprehension to generate a 3x3 matrix with values from 1 to 9.

list_c=[[1+3*i+j for j in range(0,3)] for i in range(0,3) ]
print(list_c)
```

```python
#Task 2: Dictionaries
#1.	Create a dictionary with student names as keys and their scores as values. The dictionary should have at least 5 entries

dict_stu={'Akash': 90, 'Teja': 100, 'Anshul':110, 'shivani':88, 'vimal': 99}
print(dict_stu)

#2.	Calculate the average score and print it.

sum=0
for i in dict_stu.values():
    sum=sum+i
average=sum/5
print(average)

#3.	Find and print the student(s) with the highest score
max_score=0

for i in dict_stu.values():
    if(i>max_score):
        max_score=i

students_with_max_score = [student for student, score in dict_stu.items() if score == max_score]
print(students_with_max_score)

```

```python
#Part 3: Advanced Pandas Usage
#Task 1: Complex DataFrame Manipulations
#1. Import the Pandas library.
import pandas as pd

#2. Create a DataFrame with columns 'Name', 'Age', 'Department', 'Salary', and 'Years of Service'
#with at least 7 entries.
data={ 'Name':['Akash','Vikas', 'Suvek', 'Ankit', 'Anshul','Manish','Shivam'], 'Age':[22,24,23,15,23,15,12],
      'Department':['CSE', 'IT', 'MECH', 'CSE', 'CIVIL', 'EE','IT'],
      'Salary':[22000,45000,60000,100000,55000,40000,75000],
      'YOS':[5,6,3,9,7,6,2]}
df=pd.DataFrame(data, columns=['Name', 'Department','Salary', 'YOS'])
print(df)

#3. Add a new column 'Bonus' which is 10% of the 'Salary' if 'Years of Service' is greater than 5,
#otherwise 5%


df['Bonus']=np.where(df['YOS']>5,df['Salary']*.10 , df['Salary']*.05)
df['Bonus']
```

```python
#Task 2: Handling Missing Data
#1. Introduce NaN values randomly into the DataFrame.
df.loc[1, 'Salary']=np.nan
df.loc[4,'Department']=np.nan
print(df)

#2. Fill NaN values in 'Salary' with the mean salary and drop rows where 'Department' is NaN.
df1=df['Salary'].fillna(df['Salary'].mean())
print(df1)

df1=df.dropna(subset=['Department'])
print(df1)
#3. Print the cleaned DataFrame.
```

```python
#Task 3: Grouping and Aggregation
#1. Group the DataFrame by 'Department' and calculate the average 'Salary' and total 'Years of
#Service' for each department.
gr_result=df1.groupby('Department').agg({'Salary':'mean' , 'YOS':'sum'}).reset_index()

#2. Print the grouped and aggregated data.

print(gr_result)
```

```python

#Task 1: Creating Complex Plots
#1. Import the Matplotlib library.
import matplotlib.pyplot as plt
#2. Create a scatter plot with random points. Color the points based on their distance from the
#origin.

x=np.random.rand(50)*10
y=np.random.rand(50)*10

dist= np.sqrt(x**2 + y**2)
plt.figure(figsize=(9,8))
scater_plot=plt.scatter(x,y,c=dist,cmap='viridis', s=100)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("ScatterPlot")

#3. Add a regression line to the scatter plot.
coff=np.polyfit(x,y,1)
slopp, intercept=coff
x_fitt=np.linspace(x.min(), x.max(),90)
y_fitt=slopp*x_fitt + intercept
plt.plot(x_fitt, y_fitt,color='pink', linestyle='--', linewidth=5,label='Regression Line')
plt.legend()
plt.show()
```

```python
#2. Customize the bar chart with different colors, add a legend, and annotate each bar with the
#exact value.

import matplotlib.pyplot as plt
import numpy as np


department_ = gr_result['Department']
Salaries = gr_result['Salary']


plt.figure(figsize=(8, 6))
barss = plt.bar(department_, Salaries, color='skyblue')


plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Average Salary for Each Department")


for bar, salary in zip(barss, Salaries):
    x_position = bar.get_x() + bar.get_width() / 2
    y_position = bar.get_height()
    label = f'${salary:,.0f}'
    plt.text(x_position, y_position, label, ha='center', va='bottom')


plt.legend(["Average Salary"], loc='best')


plt.show()
```

```python
#Task 3: Creating Interactive Visualizations
#1. Use Matplotlib's interactive features to create a plot where you can toggle the visibility of
#different data series.

import matplotlib.pyplot as plt
from matplotlib.widgets import CheckButtons


departments = gr_result['Department']
sal = gr_result['Salary']
sec_data = [45000, 65000, 75000, 85000]

fig, ax = plt.subplots(figsize=(8, 6))
plt.subplots_adjust(left=0.3, right=0.7, top=0.6, bottom=0.3)


bars1 = ax.bar(departments, sal, color='pink', label='sal')
bars2 = ax.bar(departments, sec_data, color='green', label='sec_data', alpha=0.6)


check = CheckButtons(plt.axes([0.04, 0.4, 0.2, 0.3]), ['Sal', 'sec_data'], [True, True])

def func(label):
    index = 0 if label == 'Sal' else 1
    bars = [bars1, bars2]
    for bar in bars:
        if index == 0 and bar == bars1:
            bar.set_visible(not bar.get_visible())
        elif index == 1 and bar == bars2:
            bar.set_visible(not bar.get_visible())
    plt.draw()

check.on_clicked(func)


ax.set_xlabel("Department")
ax.set_ylabel("Average Salary")
ax.set_title("Toggling Data")

plt.show()

#2. Implement interactivity such that clicking on a point in the scatter plot displays its
#coordinates.



x1 = [2,4,6,8,10]
y1 = [12, 17, 15, 20, 18]

fig, ax = plt.subplots(figsize=(8, 6))
sc = ax.scatter(x, y, color='red', label='Data Points')


def on_click(event):
    if event.inaxes == ax:
        x_data1, y_data1 = event.xdata, event.ydata
        print(f'Clicked at x={x_data1:.3f}, y={y_data1:.3f}')
     
        ax.annotate(f'({x_data1:.3f}, {y_data1:.3f})',
                    (x_data1, y_data1),
                    textcoords="offset points",
                    xytext=(0,10),
                    ha='center')
        plt.draw()


fig.canvas.mpl_connect('button_press_event', on_click)


ax.set_xlabel("x-axis")
ax.set_ylabel("y-Axis")
ax.set_title("scatter plot")

plt.legend()
plt.show()
```
