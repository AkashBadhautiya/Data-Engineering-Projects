```python
dataframe=pd.read_csv('Data required for the assignmnet.csv') # reading the csv file
dataframe.head(10) # printing the first 5 rows
```

```python
dataframe=dataframe.fillna(dataframe.mode()) # filling the null values with mode
print(dataframe.head()) # printing the first 5 rows
print(dataframe.shape) # printing the rows , column 
```

```python
dataframe_sorted=dataframe.sort_values(by='timestamp') #sorting the value 
print(dataframe_sorted.head()) #printing first 5 rows
print(dataframe_sorted.describe()) #printing the description
```

```python
class Segment_Tree: #initializing the class for segment tree
  def __init__(self,dataset):
    self.dataset=dataset
    self.n=len(dataset) #length of dataset
    self.tree=[{'sum': 0, 'max': float('-inf'), 'min': float('inf')}]*(2*self.n) #initializing the all nodes at leaves
    self.build_tree(dataset) #calling build dataset


  def build_tree(self,dataset):  # defining the build tree
    if not dataset: #checking the dataset for exit or not
      return
    for i in range(self.n): 
      self.tree[self.n+i]={'sum':dataset[i],'max':dataset[i],'min':dataset[i]}
    for i in range(self.n-1,0,-1):
      self.tree[i]={'sum':self.tree[2*i]['sum']+self.tree[2*i+1]['sum'],
                        'max':max(self.tree[2*i]['max'],self.tree[2*i+1]['max']),
                        'min':min(self.tree[2*i]['min'],self.tree[2*i+1]['min'])}
        #storing sum, max, min as dictionary  in tree


  def query(self,l,r): # defining query dataset
    l+=self.n # length of left
    r+=self.n #length of right
    left=l # copying the l
    right=r #coping the r
    query={'sum':0,'max':float('-inf'),'min':float('inf')} # initializing the dictionary
    while l<=r: #checking the l with r
      if l%2==1: # checking for left
        query['sum']+=self.tree[l]['sum'] if isinstance(self.tree[l],dict) else self.tree[l] #updating the sum
        query['max']=max(query['max'],self.tree[l]['max']) if isinstance(self.tree[l],dict) else max(query['max'],self.tree[l]) #updating the max
        query['min']=min(query['min'],self.tree[l]['min']) if isinstance(self.tree[l],dict) else min(query['min'],self.tree[l]) #updating the min
        l+=1 # increament of l

      if r%2==1: #checkin of left
        r-=1 #decreamenting the r by 1
        query['sum']+=self.tree[r]['sum'] if isinstance(self.tree[r],dict) else self.tree[r] #updating the sum
        query['max']=max(query['max'],self.tree[r]['max']) if isinstance(self.tree[r],dict) else max(query['max'],self.tree[r]) #updating the max
        query['min']=min(query['min'],self.tree[r]['min']) if isinstance(self.tree[r],dict) else min(query['min'],self.tree[r]) #updating the min

      l=l//2 # decreamenting the l to l/2
      r=r//2 # decreamenting the r to r/2
    query['mean']=query['sum']/(right-left+1) if right-left>1 else 0 # calculating the maen and updating 

    return query #returning the query

  def update(self,index,value): # initializing the update function
    index+=self.n #index will be index+n
    self.tree[index]=value #updating the val of tree ar index

    ind=index
    while(ind>1): #initializing the while loop , and run until ind >1
      ind=ind//2 #updating the index to ind/2
      self.tree[ind]={'sum':self.tree[2*ind]['sum']+self.tree[2*ind+1]['sum'],
                      'min':min(self.tree[2*ind], self.tree[2*ind+1]),
                      'max':max(self.tree[2*ind], self.tree[2*ind+1])}#updating the sum, min, max
```

## Assigning the time interval

```python
#defining the start_time
start_time = 1360009535

#defining the end_time
end_time = 1360009677
```

```python
#extracting the data between the given range
fil_data=dataframe_sorted[(dataframe_sorted['timestamp']>=start_time) & (dataframe_sorted['timestamp']<=end_time)]
```

```python
if not fil_data.empty: #checking the data for exits or not
  mum_data=fil_data['Mumb'].tolist() # converting to list data
  chennai_data=fil_data['Chennai'].tolist() #converting to list data
  khar_data=fil_data['Khar'].tolist()#converting to list data
  delhi_data=fil_data['Delh'].tolist() #converting to list data

  m_segment=Segment_Tree(mum_data) #calling the segment tree for mum data
  c_segment=Segment_Tree(chennai_data) #calling the segment tree for chennai data
  k_segment=Segment_Tree(khar_data) #calling the segment tree for khar data
  d_segment=Segment_Tree(delhi_data) #calling the segment tree for delhi data

  l,r=0,len(fil_data)-1 # initializing the l=0 , r=0 and n-1
  mumbai_stats=m_segment.query(l,r) #getting the query for mumbai data
  chennai_stats=c_segment.query(l,r) # getting the query for chennai data
  kharagpur_stats=k_segment.query(l,r) # getting the query for khar
  delhi_stats=d_segment.query(l,r)  #getting the query for delhi data

```
