### Query 1: Find all people who live in a specific city and work for a specific company.

```python
def query_1(tx):
    result = tx.run("""
        MATCH (p:Person)-[:WORKS_FOR]->(c:Company {name: 'InnoTech'}),
              (p)-[:LIVES_IN]->(ci:City {name: 'Boston'})
        RETURN p.name
    """)
    return [record["p.name"] for record in result]

# Execute Query 1
with driver.session() as session:
    employees_in_city = session.execute_read(query_1)
    print("Employees in Boston working for InnoTech:", employees_in_city)
```

Query 2: Find events that were hosted by companies located in a specific city.

```python
def query_2(tx):
    result = tx.run("""
        MATCH (e:Event)-[:HOSTED_BY]->(c:Company)-[:LOCATED_IN]->(ci:City {name: 'Los Angeles'})
        RETURN e.name AS event_name, c.name AS company_name
    """)
    return [{"event_name": record["event_name"], "company_name": record["company_name"]} for record in result]

# Execute Query 2
with driver.session() as session:
    events_in_city = session.execute_read(query_2)
    print("Events hosted by companies in Los Angeles:", events_in_city)
```

Query 3: Find people who have skills related to "Skill 1" or "Skill 2".

```python
def query_3(tx):
    result = tx.run("""
        MATCH (p:Person)-[:HAS_SKILL]->(s:Skill)
        WHERE s.name IN ['Machine Learning', 'Cloud Computing']
        RETURN p.name AS person_name, s.name AS skill_name
    """)
    return [{"person_name": record["person_name"], "skill_name": record["skill_name"]} for record in result]

# Execute Query 3
with driver.session() as session:
    people_with_skills = session.execute_read(query_3)
    print("People with skills related to Machine Learning or Cloud Computing:", people_with_skills)
```

Query 4: Find pairs of people who know each other and attended the same event.

```python
def query_4(tx):
    result = tx.run("""
        MATCH (p1:Person)-[:KNOWS]->(p2:Person),
              (p1)-[:ATTENDED]->(e:Event),
              (p2)-[:ATTENDED]->(e)
        RETURN p1.name AS person1, p2.name AS person2, e.name AS event_name
    """)
    return [{"person1": record["person1"], "person2": record["person2"], "event_name": record["event_name"]} for record in result]

# Execute Query 4
with driver.session() as session:
    attending_pairs = session.execute_read(query_4)
    print("Pairs of people who know each other and attended the same event:", attending_pairs)
```

#### Delete Specific Relationships

```python
def delete_specific_relationships(tx):
    tx.run("MATCH ()-[r:KNOWS]-() DELETE r")

with driver.session() as session:
    session.execute_write(delete_specific_relationships)
```

#### Delete all relationships


```python
def delete_all_relationships(tx):
    tx.run("MATCH ()-[r]-() DELETE r")

with driver.session() as session:
    session.execute_write(delete_all_relationships)
```

#### Code to Delete everything

```python
def delete_all(tx):
    tx.run("MATCH (n) DETACH DELETE n")

with driver.session() as session:
    session.execute_write(delete_all)
```
